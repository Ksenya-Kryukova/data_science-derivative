Уровень 1
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
def f(x) :
    return (np.sin(2*x+1))**5
```

```python
x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative_1')
plt.plot(x, f2x, label='derivative_2')
plt.legend()
plt.grid()
plt.show()
```
